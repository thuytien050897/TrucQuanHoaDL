```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.randn(100,1)

plt.plot(values, color = 'teal')

plt.xticks(fontsize =10)
plt.yticks(fontsize =10)
plt.xlabel("step", fontsize = 15)
plt.ylabel("value", fontsize = 15)

plt.title("Example 1", fontsize = 15)

plt.savefig('example1.pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/DV-course/sub_data.csv")
years = data.Year
values = data.Value

plt.figure(figsize=(10,3))
plt.plot(years, values, color = 'teal', linewidth = 2)

plt.xticks(np.arange(min(years), max(years)+1, 1))
plt.xticks(rotation = 45)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Value", fontsize = 15)

plt.title("Example 2", fontsize = 15)

plt.savefig('example2.pdf')
plt.show()
```

```python

#assignment
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/DV-course/tomato-yields.csv")

countries = ['United States', 'Spain', 'China', 'India']
filtered_data = data[data['Entity'].isin(countries) & (data['Year'] >= 1961) & (data['Year'] <= 2020)]

plt.figure(figsize=(10, 6))

for country in countries:
    country_data = filtered_data[filtered_data['Entity'] == country]
    plt.plot(country_data['Year'], country_data['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'], label=country)

plt.title('Tomato Yields (1961-2020)')
plt.xlabel('Year')
plt.ylabel('Tomato Yields (tonnes per hectare)')
plt.legend()
plt.grid(True)
plt.show()
```
