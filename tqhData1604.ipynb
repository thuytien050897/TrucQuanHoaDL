```python
import numpy as np
import matplotlib.pyplot as plt

height = [3,12,5,18,45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color = "teal")

plt.xticks(y_pos, bars)

plt.xticks(fontsize =15)
plt.yticks(fontsize =15)

plt.title("Example 1", fontsize=15)

plt.savefig('example1.pdf')
plt.show()
```

```python
barWidth = 0.25

bars1 = [12,30,1,8,22]
bars2 = [28,6,16,5,10]
bars3 = [29,3,24,25,17]

r1 = np.arange(len(bars1))
r2 = [x+barWidth for x in r1]
r3 = [x+barWidth for x in r2]

plt.bar(r1, bars1, color='lightseagreen',
        width = barWidth, edgecolor = 'white', label = 'var1')
plt.bar(r2, bars2, color='teal',
        width = barWidth, edgecolor = 'white', label = 'var2')
plt.bar(r3, bars3, color='darkslategrey',
        width = barWidth, edgecolor = 'white', label = 'var3')

plt.xlabel('group', fontweight = 'bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['A', 'B', 'C', 'D', 'E'])
plt.xticks(fontsize =15)
plt.yticks(fontsize =15)

plt.title("Example 2", fontsize=15)
plt.legend()

plt.savefig('example2.pdf')
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

r=[0,1,2,3,4]

raw_data = {'Bars1':[20,1.5,7,10,5],
            'Bars2':[5,15,5,10,15],
            'Bars3':[2,15,18,5,10]}
df = pd.DataFrame(raw_data)

totals = [i+j+k for i,j,k in zip(df['Bars1'], df['Bars2'], df['Bars3'])]

greenBars = [i/j*100 for i, j in zip(df['Bars1'], totals)]
orangeBars = [i/j*100 for i, j in zip(df['Bars2'], totals)]
blueBars = [i/j*100 for i, j in zip(df['Bars3'], totals)]

barWidth = 0.85
names = ('A', 'B', 'C', 'D', 'E')

plt.bar(r, greenBars, color='darkslategrey', edgecolor = 'white',
        width = barWidth)
plt.bar(r, orangeBars, bottom=greenBars, color='teal',
        width = barWidth, edgecolor = 'white')
plt.bar(r, blueBars, bottom=[i+j for i,j in zip(greenBars, orangeBars)], color='lightseagreen',
        width = barWidth, edgecolor = 'white')

plt.xticks(r, names)
plt.xlabel("group", fontsize =15)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.title("Example 4", fontsize =15)
plt.ylabel("Percentage (%)", fontsize =15)

plt.savefig("example4.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("7_OneCatOneNum.csv")

data_sorted = data.sort_values(by="Value", ascending=False) # giảm dần
top_10_exporters = data_sorted.head(10)

# Tổng hợp dữ liệu của các quốc gia còn lại thành một quốc gia 'Others'
other_countries_value = data_sorted.iloc[10:]['Value'].sum() #truy cập dòng từ hàng thứ 10 trở đi (tính từ 0)
other_countries = pd.DataFrame({"Country": ["Others"], "Value": [other_countries_value]})

# Kết hợp dữ liệu của 10 quốc gia lớn nhất và quốc gia 'Others'
df = pd.concat([top_10_exporters, other_countries])

plt.bar(df['Country'], df['Value'], color = "teal", width = 0.8)

plt.ylabel("Weapon quantity", fontsize = 15)
plt.xticks(rotation = 80, fontsize =5)
plt.yticks(np.arange(0,13000,4000), fontsize =15)

plt.title("Example 5", fontsize =15)
plt.tight_layout()
plt.savefig("example5.pdf")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tập tin CSV
data = pd.read_csv("7_OneCatOneNum.csv")

# Xóa các dòng có giá trị rỗng trong cột 'Value'
data = data.dropna(subset=['Value'])

# Chuyển cột 'Value' sang kiểu số nguyên
data['Value'] = pd.to_numeric(data['Value'])

# Sắp xếp dữ liệu theo cột 'Value'
data_sorted = data.sort_values(by="Value", ascending=False)

# Lấy 10 quốc gia có số lượng vũ khí xuất khẩu nhiều nhất
top_10_exporters = data_sorted.head(10)

# Barchart
plt.figure(figsize=(10, 6))
plt.barh(top_10_exporters["Country"], top_10_exporters["Value"], color='skyblue')
plt.xlabel('Weapons Export (billions)')
plt.ylabel('Country')
plt.title('Top 10 Weapon Exporters')
plt.gca().invert_yaxis()  # Đảo ngược trục y để hiển thị doanh nghiệp hàng đầu ở trên cùng
plt.show()

# Lolipop chart
plt.figure(figsize=(10, 6))
plt.stem(top_10_exporters["Country"], top_10_exporters["Value"], markerfmt='o', linefmt='skyblue', basefmt=" ", use_line_collection=True)
plt.xlabel('Country')
plt.ylabel('Weapons Export (billions)')
plt.title('Top 10 Weapon Exporters')
plt.xticks(rotation=45)  # Xoay nhãn trục x để dễ đọc hơn
plt.show()
```
